```python
import json
import time

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter
```

```python
df = pd.read_csv("clean_health_care_charities.csv", parse_dates=["0.Reporting Date", "1.Reporting Date", "2.Reporting Date", "3.Reporting Date", "4.Reporting Date"])
```

## Feature Extraction

### From Revenue

```python
for i in range(5):
    column_name = f"{i}.Revenue"
    new_column_name = f"{i}.Total Revenue"
    total_revenues = []
    for revenue in list(df[column_name]):
        total_revenues.append(float(revenue.split("'Total revenue': ")[1].strip("}")))
    df[new_column_name] = total_revenues
```

### From Expenses

```python
for i in range(5):
    column_name = f"{i}.Expenses"
    new_column_name = f"{i}.Total Expenses"
    total_expenses = []
    for expenses in list(df[column_name]):
        total_expenses.append(float(expenses.split("'Total expenses': ")[1].strip("}")))
    df[new_column_name] = total_expenses
```

### From Total Compensation

```python
for i in range(5):
    column_name = f"{i}.Total Compensation"
    tot_comp_name = f"{i}.Total Compensation Value"
    full_time_name = f"{i}.Full Time Employees"
    part_time_name = f"{i}.Part Time Employees"
    consulting_name = f"{i}.Consulting Fees"
    total_compensation = []
    full_time_employees = []
    part_time_employees = []
    consulting_fees = []
    for tot_comp in list(df[column_name]):
        total_compensation.append(float(tot_comp.split("'Total expenses': ")[1].strip("}")))
        full_time_employees.append(float(tot_comp.split("'Total expenses': ")[1].strip("}")))
        part_time_employees.append(float(tot_comp.split("'Total expenses': ")[1].strip("}")))
        consulting_fees.append(float(tot_comp.split("'Total expenses': ")[1].strip("}")))
    df[tot_comp_name] = total_compensation
    df[full_time_name] = full_time_employees
    df[part_time_name] = part_time_employees
    df[consulting_name] = consulting_fees
```
